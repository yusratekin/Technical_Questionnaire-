```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the data set
data = {
    'Area of Interest': ['Work', 'Family', 'Homeworks', 'Individual', 'Socializing', 'Spare Time', 'Sleep'],
    'Charles': [8.5, 0.5, 3.0, 1.0, 1.5, 3.0, 6.5],
    'Henry': [9.5, 1.0, 2.0, 1.5, 0.5, 2.5, 7.0],
    'Susan': [7.0, 1.5, 1.0, 2.5, 2.0, 2.0, 8.0]
}

df = pd.DataFrame(data)

# Creating the line plot
plt.figure(figsize=(10, 6))

# Line plot for each person
plt.plot(df['Area of Interest'], df['Charles'], marker='o', label='Charles')
plt.plot(df['Area of Interest'], df['Henry'], marker='o', label='Henry')
plt.plot(df['Area of Interest'], df['Susan'], marker='o', label='Susan')

# Plot title and labels
plt.title('Time Spent by Three Individuals on Areas of Interest Throughout the Day')
plt.xlabel('Areas of Interest')
plt.ylabel('Hours')
plt.legend(title='Individuals')

# Show
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.tight_layout()  # Adjust plot layout
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = {
    'Area of Interest': ['Work', 'Family', 'Homeworks', 'Individual', 'Socializing', 'Spare Time', 'Sleep'],
    'Charles': [8.5, 0.5, 3.0, 1.0, 1.5, 3.0, 6.5],
    'Henry': [9.5, 1.0, 2.0, 1.5, 0.5, 2.5, 7.0],
    'Susan': [7.0, 1.5, 1.0, 2.5, 2.0, 2.0, 8.0]
}

df = pd.DataFrame(data)

# Set the 'Area of Interest' column of the DataFrame as the index
df.set_index('Area of Interest', inplace=True)

# Create a stacked bar chart
df.plot(kind='bar', stacked=True, figsize=(10, 7))

# Chart title and labels
plt.title('Time Spent on Areas of Interest Throughout the Day')
plt.xlabel('Areas of Interest')
plt.ylabel('Hours')
plt.legend(title='Persons')

# Show the plot
plt.show()
```
